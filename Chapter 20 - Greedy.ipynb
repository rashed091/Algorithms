<p>In an algorithm design there is no one 'silver bullet' that is a cure for all computation problems. Different problems require the use of different kinds of techniques. A good programmer uses all these techniques based on the type of problem. 
Some commonly-used techniques are:</p>
<ol>
<li>Divide and conquer</li>
<li>Randomized algorithms</li>
<li>Greedy algorithms (This is not an algorithm, it is a <strong>technique</strong>.)</li>
<li>Dynamic programming</li>
</ol>
<p><strong>What is a 'Greedy algorithm'?</strong></p>
<p>A greedy algorithm, as the name suggests, <strong>always makes the choice that seems to be the best at that moment</strong>. This means that it makes a locally-optimal choice in the hope that this choice will lead to a globally-optimal solution. </p>
<p>How do you decide which choice is optimal?</p>
<p>Assume that you have an <strong>objective function</strong> that needs to be optimized (either maximized or minimized) at a given point. A Greedy algorithm makes greedy choices at each step to ensure that the objective function is optimized. The Greedy algorithm has only one shot to compute the optimal solution so that <strong>it never goes back and reverses the decision</strong>. </p>
<p>Greedy algorithms have some advantages and disadvantages:</p>
<ol>
<li>It is quite easy to <strong>come up with a greedy algorithm</strong> (or even
    multiple greedy algorithms) for a problem.</li>
<li><strong>Analyzing the run time for greedy algorithms will generally be much
    easier</strong> than for other techniques (like Divide and conquer). For the
    Divide and conquer technique, it is not clear whether the technique is fast or
    slow. This is because at each level of recursion the size of gets smaller and the number of
    sub-problems increases.</li>
<li>The difficult part is that for greedy algorithms <strong>you have to work much harder to understand correctness issues</strong>. Even with the correct algorithm, it is hard to prove why it is correct. Proving that a greedy algorithm is correct is more of an art than a science. It involves a lot of creativity.</li>
</ol>
<p><strong>Note:</strong> Most greedy algorithms are <strong>not</strong> correct. An example is described later in this article.</p>
<p><strong>C. How to create a Greedy Algorithm?</strong></p>
<p>Being a very busy person, you have exactly T time to do some interesting things and you want to do maximum such things. </p>
<p>You are given an array <strong>A</strong> of integers, where each element indicates the time a thing takes for completion. You want to calculate the maximum number of things that you can do in the limited time that you have.</p>
<p>This is a simple Greedy-algorithm problem. In each iteration, you have to greedily select the things which will take the minimum amount of time to complete while maintaining two variables <strong>currentTime</strong> and <strong>numberOfThings</strong>.  To complete the calculation, you must:</p>
<ol>
<li>Sort the array <strong>A</strong> in a non-decreasing order.</li>
<li>Select each to-do item one-by-one.</li>
<li>Add the time that it will take to complete that to-do item into <strong>currentTime</strong>.</li>
<li>Add one to <strong>numberOfThings</strong>.</li>
</ol>
<p>Repeat this as long as the <strong>currentTime</strong> is less than or equal to <strong>T</strong>.</p>
<p>Let <strong>A = {5, 3, 4, 2, 1}</strong> and <strong>T = 6</strong></p>
<p>After sorting, <strong>A = {1, 2, 3, 4, 5}</strong></p>
<p>After the 1<sup>st</sup> iteration:</p>
<ul>
<li><strong>currentTime</strong> = 1</li>
<li><strong>numberOfThings</strong> = 1</li>
</ul>
<p>After the 2<sup>nd</sup> iteration:</p>
<ul>
<li><strong>currentTime</strong> is 1 + 2 = 3</li>
<li><strong>numberOfThings</strong> = 2</li>
</ul>
<p>After the 3<sup>rd</sup> iteration:</p>
<ul>
<li><strong>currentTime</strong> is 3 + 3 = 6</li>
<li><strong>numberOfThings</strong> = 3</li>
</ul>
<p>After the 4<sup>th</sup> iteration, <strong>currentTime</strong> is 6 + 4 = 10, which is greater than T. 
Therefore, the answer is 3.  </p>
<p><strong>Implementation</strong></p>

<p>This example is very trivial and as soon as you read the problem, it is apparent that you can apply the Greedy algorithm to it.</p>
<p>Consider a more difficult problem-the <strong>Scheduling</strong> problem. </p>
<p>You have the following:</p>
<ul>
<li>List of all the tasks that you need to complete today</li>
<li>Time that is required to complete each task</li>
<li>Priority (or weight ) to each work.</li>
</ul>
<p>You need to determine in what order you should complete the tasks to get the most optimum result.</p>
<p>To solve this problem you need to analyze your inputs. In this problem, your inputs are as follows:</p>
<ul>
<li>Integer N for the number of jobs you want to complete</li>
<li>Lists P: Priority (or weight)</li>
<li>List T: Time that is required to complete a task</li>
</ul>
<p>To understand what criteria to optimize, you must determine the total time that is required to complete each task.</p>
<p><strong>C(j) = T[1] + T[2] + .... + T[j] where 1 &lt;= j &lt;= N</strong></p>
<p>This is because <strong>jth</strong> work has to wait till the first (j-1) tasks are completed after which it requires T[j] time for completion.</p>
<p>For example, if T = {1, 2, 3}, the completion time will be: </p>
<ul>
<li>C(1) = T[1] = 1</li>
<li>C(2) = T[1] + T[2] = 1 + 2 = 3</li>
<li>C(3) = T[1] + T[2] + T[3] = 1 + 2 + 3 = 6</li>
</ul>
<p>You obviously want completion times to be as short as possible. But it's not that simple.</p>
<p>In a given sequence, the jobs that are queued up at the beginning have a shorter completion time and jobs that are queued up towards the end have longer completion times. </p>
<p>What is the optimal way to complete the tasks? </p>
<p>This depends on your objective function. While there are many objective functions in the "Scheduling" problem, your objective function <strong>F</strong> is the <strong>weighted sum of the completion times</strong>.</p>
<p><strong>F = P[1] * C(1) + P[2] * C(2) + ...... + P[N] * C(N)</strong></p>
<p>This objective function must be minimized.</p>
<p><strong><em>Special cases</em></strong></p>
<p>Consider the special cases that is reasonably intuitive about what the optimal thing to do is. Looking at these special cases will bring forth a couple of natural greedy algorithms after which you will have to figure out how to narrow these down to just one candidate, which you will prove to be correct.</p>
<p>The two special cases are as follows:</p>
<ol>
<li>
<p>If the time required to complete different tasks is the same i.e. <strong>T[i] = T[j] where 1 &lt;= i, j &lt;= N</strong>, but they have different priorities then in what order will it make sense to schedule the jobs?</p>
</li>
<li>
<p>If the priorities of different tasks are the same i.e. <strong>P[i] = P[j] where 1 &lt;= i, j &lt;= N</strong> but they have different lengths then in what order do you think we must schedule the jobs?</p>
</li>
</ol>
<p>If the time required to complete different tasks is the same, then you should give preference to the task with the higher priority.</p>
<p><strong>Case 1</strong></p>
<p>Consider the objective function that you need to minimize. Assume that the time required to complete the different tasks is <strong>t</strong>.</p>
<p>T[i] = t where 1 &lt;= i &lt;= N</p>
<p>Irrespective of what sequence is used, the completion time for each task will be as follows:<br />
C(1) = T[1] = t<br />
C(2) = T[1] + T[2] = 2 * t<br />
C(3) = T[1] + T[2] + T[3] = 3 * t<br />
...<br />
C(N) = N * t  </p>
<p>To make the objective function as small as possible the highest priority must be associated with the shortest completion time.</p>
<p><strong>Case 2</strong></p>
<p>In the second case, if the priorities of different tasks are the same, then you must favor the task that requires the least amount of time to complete. Assume that the priorities of the different tasks is <strong>p</strong>.</p>
<p>F = P[1] * C(1) + P[2] * C(2) + ...... + P[N] * C(N)<br />
F = p * C(1) + p * C(2) + ...... + p * C(N)<br />
F = p * (C(1) + C(2) + ...... + C(N))  </p>
<p>To minimize the value of F, you must minimize (C(1) + C(2) + ...... + C(N)), which can be done if you start working on the tasks that require the shortest time to complete. </p>
<p>There are two rules. Give preference to tasks that:</p>
<ul>
<li>Have a higher priority</li>
<li>Take less time to complete</li>
</ul>
<p>The next step is to move beyond the special cases, to the general case. In this case, the priorities and the time required for each task are different. </p>
<p>If you have 2 tasks and both these rules give you the same advice, then the task that has a higher priority and takes less time to complete is clearly the task that must be completed first. But what if both these rules give you conflicting advice? What if you have a pair of tasks where one of them has a higher priority and the other one requires a longer time to complete? ( i.e. P[i] &gt; P[j] but T[i] &gt; T[j] ). Which one should you complete first?</p>
<p>Can you aggregate these 2 parameters (time and priority) into a single score such that if you sort the jobs from higher score to lower score you will always get an optimal solution?</p>
<p>Remember the 2 rules. </p>
<ol>
<li>Give preference to higher priorities so that the <strong>higher priorities lead to a higher score</strong>.</li>
<li>Give preference to tasks that require less time to complete so that the <strong>more time that is required should decrease the score</strong>.</li>
</ol>
<p>You can use a simple mathematical function, which takes 2 numbers (priority and time required) as the input and returns a single number (score) as output while meeting these two properties. (There are infinite number of such functions.)</p>
<p>Lets take two of the simplest functions that have these properties</p>
<ul>
<li><strong>Algorithm #1</strong>: order the jobs by <strong>decreasing value of ( P[i] - T[i] )</strong></li>
<li><strong>Algorithm #2</strong>: order the jobs by <strong>decreasing value of ( P[i] / T[i] )</strong></li>
</ul>
<p>For simplicity we are assuming that there are <strong>no ties.</strong></p>
<p>Now you have two algorithms and at least one of them is wrong. Rule out the algorithm that does not do the right thing.</p>
<p>T = {5, 2} and P = {3, 1}  </p>
<p>According to the algorithm #1 ( P[1] - T[1] ) &lt; ( P[2] - T[2] ), therefore, the second task should be completed first and your objective function will be:</p>
<p>F = P[1] * C(1) + P[2] * C(2) = 1 * 2 + 3 * 7 = 23  </p>
<p>According to algorithm #2 ( P[1] / T[1] ) &gt; ( P[2] / T[2] ), therefore, the first task should be completed first and your objective function will be:</p>
<p>F = P[1] * C(1) + P[2] * C(2) = 3 * 5 + 1 * 7 = 22  </p>
<p>Algorithm #1 will not give you the optimal answer and, therefore, algorithm #1 is not (<strong>always</strong> ) correct. </p>
<p><strong>Note</strong>: Remember that Greedy algorithms are often <strong>WRONG</strong>. Just because algorithm #1 is not correct, it does not imply that algorithm #2 is guaranteed to be correct. It does, however, turn out that in this case algorithm #2 is always correct. </p>
<p>Therefore, the final algorithm that returns the optimal value of the objective function is:</p>

<p><strong><em>Time complexity</em></strong>
You have 2 loops taking O(N) time each and one sorting function taking O(N * logN). Therefore, the overall time complexity is O(2 * N + N * logN) = <strong>O(N * logN)</strong>. </p>
<p><strong>Proof of Correctness</strong></p>
<p>To prove that <strong>algorithm #2</strong> is correct, use <strong>proof by contradiction</strong>. Assume that what you are trying to prove is false and from that derive something that is obviously false.</p>
<p>Therefore, assume that this greedy algorithm does not output an optimal solution and there is another solution (not output by greedy algorithm) that is better than greedy algorithm.</p>
<p><strong>A = Greedy schedule (which is not an optimal schedule)</strong><br />
<strong>B = Optimal Schedule (best schedule that you can make)</strong></p>
<p><strong>Assumption #1:</strong> all the ( P[i] / T[i] ) are <strong>different</strong>.<br />
<strong>Assumption #2:</strong> (just for simplicity, will not affect the generality) <strong>( P[1] / T[1] ) &gt; ( P[2] / T[2] ) &gt; .... &gt; ( P[N] / T[N] )</strong> </p>
<p>Because of assumption #2, the greedy schedule will be <strong>A = ( 1, 2, 3, ....., N )</strong>. Since A is not optimal (as we considered above) and A is not equal to B (because B is optimal), you can claim that <strong>B must contain two consecutive jobs ( i, j ) such that the earlier of those 2 consecutive jobs has a larger index ( i &gt; j ).</strong> This is true because the only schedule that has the property, in which the indices only go up, is A = ( 1, 2, 3, ...., N ).</p>
<p>Therefore, <strong>B = ( 1, 2, ..., i, j, ... , N )  where i &gt; j</strong>.</p>
<p>You also have to think about what is the profit or loss impact if you swap these 2 jobs. Think about the effect of this swap on the completion times of the following:</p>
<ol>
<li>Work on k other than i and j  </li>
<li>Work on i  </li>
<li>Work on j  </li>
</ol>
<p>For k, there will be 2 cases:</p>
<p><strong>When k is on the left of i and j in B</strong>
If you swap i and j, then there will be no effect on the completion time of k. </p>
<p><strong>When k is on the right of i and j in B</strong>
After swapping, the completion time of k is C(k) = T[1] + T[2] + .. + T[j] + T[i] + .. T[k], k will remain same.</p>
<p>For i the completion time:
Before swapping was C(i) = T[1] + T[2] + ... + T[i]
After swapping is C(i) = T[1] + T[2] + ... + T[j] + T[i]</p>
<p>Clearly, the completion time for i goes up by T[j] and the completion time for j goes down by T[i].</p>
<p><strong>Loss due to the swap is (P[i] * T[j])</strong><br />
<strong>Profit due to the swap is (P[j] * T[i])</strong> </p>
<p>Using assumption #2, <strong>i &gt; j implies that ( P[i] / T[i] ) &lt; ( P[j] / T[j] )</strong>. Therefore <strong>( P[i] * T[j] ) &lt; ( P[j] * T[i] ) which means Loss &lt; Profit</strong>. This means that <strong>swap improves B but it is a contradiction</strong> as we assumed that B is the optimal schedule. This completes our proof.</p>
<p><strong>Where to use Greedy algorithms?</strong></p>
<p>A problem must comprise these two components for a greedy algorithm to work:</p>
<ol>
<li>
<p>It has <strong>optimal substructures</strong>. The optimal solution for the problem contains optimal solutions to the sub-problems.</p>
</li>
<li>
<p>It has a <strong>greedy property</strong> (hard to prove its correctness!). If you make a choice that seems the best at the moment and solve the remaining sub-problems later, you still reach an optimal solution. You will never have to reconsider your earlier choices.</p>
</li>
</ol>
